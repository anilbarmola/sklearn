### Problem
You need to rescale the values of a numerical feature to be between two values.
### Solution
Use scikit-learnâ€™s MinMaxScaler to rescale a feature array

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
feature=np.array([[100.1],[-500.2],[102.5],[0],[101.2],[300.2]])
```

```python
plt.plot(feature)
```

Applying min_max scaller

```python
#creat scaler
minmax_scaler=preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
#scale feature
scale_feature=minmax_scaler.fit_transform(feature)
```

```python
scale_feature
```

```python
plt.plot(scale_feature)
```

### Graph before scalling and after scalling

```python
plt.subplot(1,2,2)
plt.plot(feature)
plt.title("Before sclaed")

plt.subplot(1,2,1)
plt.plot(scale_feature)
plt.title("After sclaed")
```
